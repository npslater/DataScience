## Data Fitting

Here are some examples I use to demonstrate how both classical ML and deep learning techniques can be used to fit functions to both linear and non-linear data.

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import sklearn.datasets
import sklearn.linear_model
from sklearn.metrics import confusion_matrix, mean_squared_error, r2_score
#from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import StandardScaler
import pandas as pd
import itertools
import math
import pandas as pd

%matplotlib inline
```

```python
class ConfusionMatrix:
    
    def __init__(self, y, yhat):
        self.y = y
        self.yhat = yhat
        self.cm = confusion_matrix(y, yhat)
        
    def accuracy(self):
        tn,fp,fn,tp = self.cm.ravel()
        try:
            return float(tn+tp)/float(tn+fp+fn+tp)
        except ZeroDivisionError:
            return 0.0
    
    def precision(self):
        tn,fp,fn,tp = self.cm.ravel()
        try:
            return float(tp)/float(tp+fp)
        except ZeroDivisionError:
            return 0.0
    
    def recall(self):
        tn,fp,fn,tp = self.cm.ravel()
        try:
            return float(tp)/float(tp+fn)
        except ZeroDivisionError:
            return 0.0
    
    def F1(self):
        
        p = self.precision()
        r = self.recall()
        try:
            return 2.0 * ((p*r)/(p+r))
        except ZeroDivisionError:
            return 0.0
    
    def print_performance(self):
        print('Accuracy: {:0.2f}%'.format(self.accuracy()*100))
        print('Precision: {:0.2f}%'.format(self.precision()*100))
        print('Recall: {:0.2f}%'.format(self.recall()*100))
        print('F1: {:0.2f}'.format(self.F1()))

    def plot(self,
             classes,
             normalize=False,
             title='Confusion matrix',
             cmap=plt.cm.Blues):
        """
        This function prints and plots the confusion matrix.
        Normalization can be applied by setting `normalize=True`.
        """
        if normalize:
            cm = self.cm.astype('float') / self.cm.sum(axis=1)[:, np.newaxis]
            print("Normalized confusion matrix")
        else:
            print('Confusion matrix, without normalization')

        print(cm)

        plt.imshow(cm, interpolation='nearest', cmap=cmap)
        plt.title(title)
        plt.colorbar()
        tick_marks = np.arange(len(classes))
        plt.xticks(tick_marks, classes, rotation=45)
        plt.yticks(tick_marks, classes)

        fmt = '.2f' if normalize else 'd'
        thresh = cm.max() / 2.
        for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
            plt.text(j, i, format(cm[i, j], fmt),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")

        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.tight_layout()
        plt.savefig('confusion.png')
    
    
```

```python
def plot_decision_boundaries(X, y, model_class, **model_params):
    """Function to plot the decision boundaries of a classification model.
    This uses just the first two columns of the data for fitting 
    the model as we need to find the predicted value for every point in 
    scatter plot.
    
    One possible improvement could be to use all columns fot fitting
    and using the first 2 columns and median of all other columns
    for predicting.
    
    Adopted from:
    http://scikit-learn.org/stable/auto_examples/ensemble/plot_voting_decision_regions.html
    http://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_digits.html
    """
    reduced_data = X[:, :2]
    model = model_class(**model_params)
    model.fit(reduced_data, y)

    # Step size of the mesh. Decrease to increase the quality of the VQ.
    h = .02     # point in the mesh [x_min, m_max]x[y_min, y_max].    

    # Plot the decision boundary. For that, we will assign a color to each
    x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
    y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Obtain labels for each point in mesh using the model.
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])    

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
    plt.savefig('decision_boundary.png')
    return plt
```

### Data Sets

Define some convenience methods for creating and fetching different types of data sets.

```python
class CircularData:
    def __init__(self, radius, m):
        self.radius = radius
        self.m = m
        
    def generate(self):
        X = np.zeros((2,self.m))
        X[0,:] = np.random.uniform(0-self.radius, high=0+self.radius, size=self.m)
        X[1,:] = np.random.uniform(0-self.radius, high=0+self.radius, size=self.m)
        y = np.square(X[0,:]) + np.square(X[1,:]) < np.square(self.radius)
        return X,y
    

def housing_data(file='housing.csv', 
                 num_samples=100, 
                 features=None, 
                 train_split_pct=0.8):
    housing_df = pd.read_csv(file).dropna().sample(num_samples)
    scaler = StandardScaler()
    if features is not None:
        X = scaler.fit_transform(housing_df[features]) 
    else:
        X = scaler.fit_transform(housing_df.loc[:, housing_df.columns != 'median_house_value'])
    
    y = housing_df['median_house_value']
    test_set_size = int(num_samples - (train_split_pct*num_samples))
    return X[:-test_set_size],X[-test_set_size:],y[:-test_set_size],y[-test_set_size:]

def cancer_fatality_data(cutoff=0.8, num_samples=1000, train_split_pct=0.8):
    classes=[1,2]
    ages = np.arange(55,95)
    X = np.zeros((num_samples,1))
    y=np.zeros(X.shape[0])
    for x in range(0,num_samples):
        age = math.floor(np.random.uniform(55,95))
        index = np.squeeze(np.where(ages==age))
        X[x] = age
        if age < 65:
            prob = np.random.uniform(0,1,3)
        else:
            prob = np.random.uniform(0,1,1)
            
        if np.squeeze(np.where(prob > cutoff)).size == prob.size:
            y[x] = classes[1]
        else:
            y[x] = classes[0]
    test_set_size = int(num_samples - (train_split_pct*num_samples))
    return X[:-test_set_size],X[-test_set_size:],y[:-test_set_size],y[-test_set_size:]
        
```

### Linear data

Let's examine how to fit a linear function to data.  In this case we will be using housing price data from California.  We'll do a simple single variable linear regression of square footage against housing prices to see if square footage is a decent predictor of prices.

```python
features = [
    'longitude',
    'latitude',
    'housing_median_age',
    'total_rooms',
    'total_bedrooms',
    'population',
    'households',
    'median_income'] 

X_train,X_test,y_train,y_test = housing_data(features=features, num_samples=1000)
X_train_single = X_train[:,7].reshape(-1,1)
X_test_single = X_test[:,7].reshape(-1,1)

regr = sklearn.linear_model.LinearRegression()
regr.fit(X_train_single, y_train)
y_hat = regr.predict(X_test_single)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_hat))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_hat))

# Plot outputs
plt.scatter(X_test_single, y_test,  color='black')
#plt.scatter(X_train, y_train, color='red')
plt.plot(X_test_single, y_hat, color='blue', linewidth=3)

plt.xlabel('median income')
plt.ylabel('price')
plt.xticks(())
plt.yticks(())

plt.savefig('home_price_reg_single.png', bbox_inches='tight')
plt.show()
```

Let's see if we get a better model by using more the features in the dataset

```python
regr.fit(X_train, y_train)
y_hat = regr.predict(X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_hat))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_hat))
```

### Linear Classification

Linear regression works well when we want to predict a continuous real value, but it's not as effective when dealing with data that falls into categories.  A common ML application is to predict the category a given item falls into.  This is called 'classification' and even though one could treat this as a regression problem that predicts the values (0,1), linear regression doesn't actually work well for this.  There's a better way to handle this.

```python
X_train,X_test,y_train,y_test= cancer_fatality_data()

regr = sklearn.linear_model.LinearRegression()
regr.fit(X_train, y_train)
y_hat = regr.predict(X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_hat))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_hat))

# Plot outputs
plt.scatter(X_test, y_test,  c=y_test)
plt.plot(X_test, y_hat, color='blue', linewidth=3)

plt.xlabel('age')
plt.ylabel('survival status')
plt.savefig('cancer_survival.png')
plt.show()
```

### Logistic Regression

Linear regression doesn't do well with a classificaton problem, like the one above.  In this case, the target variable has only two discrete values - 1 and 2.  You can see that the regression line is basically predicting values right in the middle.  For classification, it's much more effective to use a different linear model, called 'logistic regression'.

```python
classifier = sklearn.linear_model.LogisticRegressionCV(solver='liblinear')
classifier.fit(X_train, y_train)
yhat = classifier.predict(X_test)
print(yhat)

cm = ConfusionMatrix(y_test, yhat)
cm.print_performance()
cm.plot(['Non-Fatal', 'Fatal'], normalize=True)
```

### Circular data

Let's generate a non-linear data set.  In this case, the dataset resembles a circle.  If we want to predict whether a point is the negative class (outisde the circle) or positive class (inside the circle) given a point's cartesian coordinates, intuitively it doesn't not seem like a linear model will work very well.  Let's find out.

```python
circular = CircularData(2,400)
X,y = circular.generate()
plt.scatter(X[0,:], X[1,:], c=y)
plt.savefig('circular.png')
```

### Fit data using linear model
Accuracy is better than a coin flip, but false positives are high.  100% of negative classes are falsely classified as positive by this linear model.

```python
classifier = sklearn.linear_model.LogisticRegressionCV(solver='liblinear')
classifier.fit(X.T, y.T)
yhat = classifier.predict(X.T)

cm = ConfusionMatrix(y, yhat)
cm.print_performance()
cm.plot(['Purple', 'Yellow'], normalize=True)
```

```python
plot_decision_boundaries(X.T, y.T, sklearn.linear_model.LogisticRegressionCV)
```

### Improve performance through feature engineering

Not surpisingly, the linear model doesn't fit this non-linear data very well.  Does this mean we have to use a differnt ML technique better suited for non-linear data?  Not necessarily.  Instead, we can use feature engineering to add polynomial terms to the linear model.  If we create two new features by squaring each of the cartesian coordinates in the training examples, we will fit this circular-shaped data set much more accurately when we re-train using logistic regression.

```python
X[0,:] = np.square(X[0,:])
X[1,:] = np.square(X[1,:])

classifier.fit(X.T, y.T)
yhat = classifier.predict(X.T)
cm = ConfusionMatrix(y, yhat) 
cm.print_performance()
cm.plot(['Purple', 'Yellow'], normalize=True)
```

### Test if the model generalizes

This looks a lot better, but any model giving near 100% accuracy may be overfitting.  Let's train a new model using our polynomial (i.e. squared) features, but test it using entirely new datasets.  For this, we will need feature normalization because the range of values in the test data sets will not be the same as the training data set.

```python
def normalize(vals):
    vals = vals - np.mean(vals)
    vals = vals/np.std(vals)
    return vals
```

```python
circular = CircularData(2,400)
train_X,train_y = circular.generate()
plt.scatter(train_X[0,:], train_X[1,:], c=train_y)
```

```python
train_X_norm = np.zeros(train_X.shape)
train_X_norm[0,:] = np.square(normalize(train_X[0,:]))
train_X_norm[1,:] = np.square(normalize(train_X[1,:]))
classifier.fit(train_X_norm.T, train_y.T)
train_yhat = classifier.predict(train_X_norm.T)

cm = ConfusionMatrix(train_y, train_yhat)
cm.print_performance()
cm.plot(['Purple', 'Yellow'], normalize=True)
```

```python
plot_decision_boundaries(train_X_norm.T, train_y.T, sklearn.linear_model.LogisticRegressionCV)
```
